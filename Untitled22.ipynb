**Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом**

```python
df_h = df.loc[df.salary == 'high', 'average_montly_hours']
df_low = df.loc[df.salary == 'low', 'average_montly_hours']
print(df_h.shape, df_low.shape)
```

```python
df_h.plot(kind = 'density')
df_low.plot(kind = 'density')
```

```python
stat, p = stats.shapiro(df_h)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
stat, p = stats.shapiro(df_low)

print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Вероятно нормальное распределение')
else:
    print('Вероятно не нормальное распределение')
```

```python
# выборки не имеют нормального распределения, размер выборок разный.
# H0 - сотрудники с высоким окладом проводят на работе стлько же времени, как и сотрудники с низким окладом
```

```python
stat, p = stats.mannwhitneyu(df_h, df_low)

print(f"Статистика = {stat:.5f}, p = {p:.5f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, распределения, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, распределения, вероятно, различаются')
```

```python
# количество времени, проводимого на работе, от зарплаты, вероятно, не зависит
```

**Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):**

● Доля сотрудников с повышением за последние 5 лет

● Средняя степень удовлетворенности

● Среднее количество проектов

```python


def summary(x):
    result = {
        'Средняя степень удовлетворенности': x['satisfaction_level'].mean(),
        'Среднее количество проектов': x['number_project'].mean(),
        'Доля сотрудников с повышением за последние 5 лет': sum(x['promotion_last_5years']*100/x['promotion_last_5years'].count())
    }
    return pd.Series(result).round(3)

res = df.groupby('left').apply(summary)

res
```

**Разделить данные на тестовую и обучающую выборки  Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary) Оценить качество модели на тестовой выборки.**

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
data = df.drop(['department','salary','left'], axis=1)

X_train, X_test, y_train, y_test = train_test_split(data, df['left'], test_size= 0.3, random_state=42)
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
accuracy_score(y_test, lda.predict(X_test)).round(5)
```
